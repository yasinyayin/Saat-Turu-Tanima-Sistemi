# Saat TÃ¼rÃ¼ TanÄ±ma Sistemi - YoloV8n Model EÄŸitimi ve Test AÅŸamasÄ± #

```python
from ultralytics import YOLO
import os

# 1) proje klasÃ¶r yolunu ayarlama
base_dir = r"D:\saat_turu_tanima"
os.chdir(base_dir)
# 2) data.yaml klasÃ¶r yolunu ayarlama
data_yaml_path = os.path.join(base_dir, "dataset", "data.yaml")
print("Ã‡alÄ±ÅŸma klasÃ¶rÃ¼ :", os.getcwd())
print("data.yaml yolu  :", data_yaml_path)
print("data.yaml var mÄ±?:", os.path.exists(data_yaml_path))
# 3) modeli yÃ¼kleme aÅŸamasÄ± (YOLOv8n)
model1 = YOLO("yolov8n.pt")
```

```python
# 4) model eÄŸitim aÅŸamasÄ± parametreler
results1 = model1.train(
    data=data_yaml_path,          # data.yaml yolu (train/val/test ve sÄ±nÄ±flar burada tanÄ±mlÄ±)
    epochs=30,                    # eÄŸitim turu sayÄ±sÄ± (60 epoch)
    imgsz=640,                    # Girdi gÃ¶rÃ¼ntÃ¼ boyutu (640x640)
    batch=8,                     # batch size (aynÄ± anda iÅŸlenecek gÃ¶rÃ¼ntÃ¼ sayÄ±sÄ±)
    device=0,                     # 0 = GPU kullan (RTX 3050) 
    project="runs_saat_tanima",   # sonuÃ§larÄ±n kaydedileceÄŸi klasÃ¶r
    name="yolov8nmodelegitim",     # eÄŸitime Ã¶zel klasÃ¶r ismi
    seed=42,                      # rastgelelik sabit (tekrar edilebilir sonuÃ§lar iÃ§in)
    patience=10,                  # 10 epoch boyunca iyileÅŸme olmazsa erken durdurma
)
print("eÄŸitim tamamlandÄ±! ğŸ‰")

```

```python
# 5) best.pt klasÃ¶r yolunu ayarlama
best_path = r"D:/saat_turu_tanima/runs_saat_tanima/yolov8nmodelegitim2/weights/best.pt"

print("best.pt var mÄ±?:", os.path.exists(best_path))

# modeli yÃ¼kle
model_exp1 = YOLO(best_path)

# test seti Ã¼zerinde deÄŸerlendirme
metrics1 = model_exp1.val(
    data=data_yaml_path,   # aynÄ± data.yaml
    split="test",           # sadece test seti
    device=0
)

print("==== TEST SONUÃ‡LARI ====")
print("mAP50-95 :", metrics1.box.map)
print("mAP50    :", metrics1.box.map50)
print("Precision:", metrics1.box.mp)
print("Recall   :", metrics1.box.mr)
```

```python
from ultralytics import YOLO

# eÄŸittiÄŸim best pt
model = YOLO(
    r"D:\saat_turu_tanima\runs_saat_tanima\yolov8nmodelegitim\weights\best.pt"
)

# canlÄ± kamera testi
model.predict(
    source=0,        # dahili kamera
    show=True,       # ekranda gÃ¶ster
    #save=True,       # Ã§Ä±ktÄ±larÄ± kaydet
    imgsz=640,       # eÄŸitimle uyumlu
    conf=0.4,        # en dengeli eÅŸik 
    device=0         # GPU (RTX 3050)
)
```
